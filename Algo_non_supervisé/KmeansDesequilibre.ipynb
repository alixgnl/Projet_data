```python
# On commence par importer les biblioth√®ques qui vont nous servir.
import numpy as np
import scipy.io as sp
import matplotlib.pyplot as plt
```

```python
# on charge notre dataset (les features X ainsi que les labels y : y vrai)
# nous n'utiliserons pas les y pour le non supervis√© mais simplement pour rem√©dier au d√©s√©quilibre de classe)

dataSet = sp.loadmat('breastw.mat') 
print((dataSet))
```

```python
# On extrait les donn√©es du champ 'X' du dictionnaire
X = dataSet['X'] 
# on regarde la taille de la matrice
np.shape(X) 
```

```python
# on extrait les donn√©es du champ 'y'
label = dataSet['y'] 
# on regarde la taille du vecteur
np.shape(label)
```

```python
# Conversion du dataset d'un dictionnaire √† un dataframe
import pandas as pd
dataset_df = pd.DataFrame(np.hstack((dataSet['X'], dataSet['y'])))
```

```python
dataset_df .columns
```

```python
X = dataset_df [[0, 1, 2, 3, 4, 5, 6, 7, 8]]
print(X)
```

```python
label = dataset_df [9]
print(label)
```

```python
import seaborn as sns
sns.pairplot(X)
```

Le graphique permet de visualiser dans des graphiques 2D les features par paire.

Pour avoir une id√©e g√©nrale de la r√©partiotion de notre dataset nous devons passer par une r√©duction de dimension de 9 √† 2 afin de visusaliser nos √©chantillons dans un graphique.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X)
```

```python
plt.scatter(x_pca[:,0], x_pca[: , 1],c=label)
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2, init ='k-means++', n_init = 10, max_iter = 300)
# La m√©thode .fit() ex√©cute le mod√®le sur les features
model.fit(X) 
# La m√©thode .predict() permet de pr√©dire pour chaque √©chantillon/observation son cluster d'appartenance et donc le centro√Øde associ√©
y_pred = model.predict(X)
centers = model.cluster_centers_
centers_pca = pca.fit_transform(centers)
plt.scatter(x_pca[:,0], x_pca[:,1], c=y_pred)
plt.scatter(centers_pca[:,0],centers_pca[:,1], c='r')
model.labels_
```

```python
np.unique(model.labels_, return_counts = True)
```

```python
y_pred = pd.DataFrame(columns= [1,2,3,4,5])
for k in range(1,6):
    model = KMeans(n_clusters= k)
    model.fit(X)
    #X[8+k] = model.labels_
    y_pred[k] = model.labels_
    print(f' Score en fonction du nb de cluster, k = {k} est :', model.score(X))
    
```

```python
y_pred
```

```python
fig, axs = plt.subplots(nrows =1, ncols = 5, figsize = (30,5))
for i, ax in enumerate(fig.axes, start =1):
    ax.scatter(x=x_pca[:, 0], y =x_pca[:, 1], c = y_pred[i])
    ax.set_title(f'Nb Cluster : {i}')
```

Avant m√™me de s√©parer notre dataset en un jeu d'entrainement et de test pour ensuite estimer les performances du mod√®le. Il faut amont remedier au d√©s√©quilibre des classes en effet nous n'avons pas la m√™me proprortion d'√©chantillons malins et d'√©chantillons b√©nins. On peut pour cela r√©aliser une cross validation ou bien un sur √©chantillonage de la classe en sosu-effectif. Nous partirons sur la premi√®re option avec stratified k fold cross validation qui permet au k partition de contenir environ la m√™me rproportion d'√©chantillon de chaque classe.

Attention : La valeur de K ne doit √™tre ni trop basse ni trop haute, et on choisit g√©n√©ralement une valeur comprise entre 5 et 10 en fonction de l‚Äôenvergure du dataset. Par exemple, si K=10, le dataset sera divis√© en 10 parties.

Une valeur K plus √©lev√©e m√®ne √† un mod√®le moins biais√©, mais une variance trop large peut conduire √† un sur-ajustement. Une valeur plus basse revient √† utiliser la m√©thode Train-Test Split.
source : https://datascientest.com/cross-validation

On va d'abord voir les r√©sultats que l'on peut obtenir avec Kmeans sans tariter le d√©s√©quilibre de classe.

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(X, label, test_size = 0.2, random_state= 2)
print('Train set :',x_train.shape)
print('Test set :',x_test.shape)
```

```python
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train, alpha=0.8)
plt.title('Train set')
plt.subplot(122)
plt.scatter(x_test_pca[:, 0], x_test_pca[:, 1],c=y_test, alpha=0.8)
plt.title('Test set')
```

On peut mainteneant utiliser notre dataset d√©coup√© pour entrainer et √©valuer notre mod√®le Kmeans

```python
model.fit(x_train) 
print('Train score :', model.score(x_train))
print('Test score :', model.score(x_test))
```
