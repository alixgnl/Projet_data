```python
dataSet = sp.loadmat('breastw.mat') #On charge le dataset
print(type(dataSet))
```

```python
X = dataSet['X'] #On extrait la donnée du champ 'X' du dictionnaire
print('dimension de X', X.shape )
Y = dataSet['y'] #On extrait les données liées à la clé 'target'
print(np.shape(Y))
print(Y)
```

```python
print('dimention de y', Y.shape)
```

```python

def perceptron_sgd(X, Y):
    w = np.zeros(len(X[0])) # initialisation du poids avec un vecteur de 0
    eta = 1 #on set le learning rate à 1
    epochs = 30 #set du nombre d'epoch

    for t in range(epochs):
        for i, x in enumerate(X):
            if (np.dot(X[i], w)*Y[i]) <= 0:
                w = w + eta*X[i]*Y[i]

    return w


w = perceptron_sgd(X,Y)
print(w)
print('dimention de w', w.shape)
```
